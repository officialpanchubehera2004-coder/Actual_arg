# 1.Positional arg.

```python
def person(name,age):
    print(name)
    print(age)             # 'nit' is position of name and 21 is position of age.

person('nit',21)
```

# 2.Keyword arg

```python
def person(name,age):
    print(name)
    print(age-5)

person(21,'nit')
```

```python
def person(name,age):
    print(name)
    print(age-5)           # we use keywords age and name.

person(age=21,name='nit')
```

# 3.Default Arg.

```python
def person(name,age=18):
    print(name)
    print(age)           # age=18 is default arg

person('jinwo sung')
```

```python
def person(name,age=18):
    print(name)
    print(age)           # because arg. is call so age 21 is print
                          # if arg. is not calling then default arg. is working.

person('jinwo sung',21)
```

# 4.Variable length arg.

```python
def sum(a,b):
    c =a+b
    print(c)
sum(5,6)    
```

```python
def sum(a,b):
    c =a+b
    print(c)
sum(5,6,7,8)  
```

```python
def sum(a,*b):
    c =a+b
    print(c)      # int and tuple we can not add
sum(5,6,7,8)
```

```python
def sum(a,*b):
    print(a)
    print(b)
    print(type(a))
    print(type(b))
sum(5,6,7,8)    
```

```python
def sum(a,*b):
    c =a
    for i in b:
        c = c+i
    print(c)
sum(5,6,7,8)    
```

```python
def sum(a,*b):
    c =0
    for i in b:
        c = c+i
    print(c)
sum(5,6,7,8)  
```
